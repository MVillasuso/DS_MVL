### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline



from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
USA_Housing = pd.read_csv('../data/USA_Housing.csv')
```

```python
USA_Housing.shape
```

```python
USA_Housing.info()
```

```python
USA_Housing.columns
```

```python
Xcols = USA_Housing[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y = USA_Housing['Price']
```

```python
y.shape
```

```python
# Regresión lineal de cada columna
y = USA_Housing['Price']
for elem in Xcols:  # Por cada una de las columnas numericas 
    X=USA_Housing[[elem]]
    # Se selecciona el subset para entrenar el modelo y para probarlo
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20) 
    # Se calcula el modelo de regresión lineal con el conjunto de entrenamiento
    lin_reg = LinearRegression()
    lin_reg.fit(X_train, y_train)
    # Gráfico de dispersión de los puntos (Valores reales)
    plt.scatter(X, y, color='b')
    #Aplicación del modelo obtenido a los valores de X (la línea roja es la función lineal obtenida que mejor representa los datos)
    plt.plot(X, lin_reg.predict(X), color='red')
    plt.title(elem)
    plt.show()
```
