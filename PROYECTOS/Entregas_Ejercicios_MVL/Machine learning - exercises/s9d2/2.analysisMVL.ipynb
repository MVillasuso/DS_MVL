¿Tendría algún sentido clasificar el nombre de los alimentos a partir del top5 de columnas numéricas con menos valores NaN?

```python
df.info()
```

```python
# Ver cuales son las columnas numéricas
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
```

```python
len(numeric_cols)
```

```python
# Elimino columnas donde todos sus elementos sean nulos (por ende NO aportan información)
df.dropna(how="all", axis=1, inplace = True)
```

```python
msno.bar(df)
```

```python
# Elimina las columnas donde haya  muy pocos valores (hasta 5 valores diferentes a null.(tresh))
df.dropna(thresh=5, axis=1, inplace=True)
```

```python
msno.bar(df)
```

```python
df.info()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
```

```python
numeric_cols
```

```python
# Porcentaje de Nan por cada columna
porcNan = df.isnull().sum()/df.shape[0]*100
```

```python
porcNan.sort_values(inplace=True)
```

Veamos el porcentaje de valores nulos en cada columna numérica 

```python
porcNan[porcNan.index.isin(numeric_cols)]
```

Al final las columnas numéricas con información se reducen a 22. 

`No creo que tenga sentido clasificar por las 5 columnas numéricas con menos valores nulos pues las mismas se refieren a valores únicos por cada fila (código, fecha de modificacion y creación)`

`Para hacer la clasificación de alimentos yo lo haría con las columnas de ingredientes con menos de la mitad de valores nulos (14 columnas en total)`
