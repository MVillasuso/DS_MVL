### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv) 

```python
dset = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv")
```

### Step 3. Assign it to a variable titanic 

```python
titanic = dset
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.info()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index("PassengerId")
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
titxsex = titanic.groupby("Sex")["PassengerId"].count().to_frame()
```

```python
titxsex
```

```python
titxsex.reset_index(inplace=True)
```

```python
titxsex
```

```python
plt.subplot(aspect=1, title=' TITANIC  - Male / Female proportion')

plt.pie(titxsex.PassengerId, labels = titxsex.Sex, autopct='%1.1f%%', shadow=True) #autopct muestra % dentro
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
graph = sns.scatterplot(x="Age", y="Fare", hue= "Sex",
            sizes=(40, 400), alpha=.5, palette="deep",
            data=titanic)
plt.title = ('Fare by Age')
```
