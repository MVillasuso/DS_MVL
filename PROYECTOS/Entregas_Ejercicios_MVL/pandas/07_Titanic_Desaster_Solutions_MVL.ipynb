### Step 5. Create a pie chart presenting the male/female proportion

```python
titxsex = titanic.groupby("Sex")["PassengerId"].count().to_frame()
```

```python
titxsex
```

```python
titxsex.reset_index(inplace=True)
```

```python
titxsex
```

```python
titxsex.PassengerId
```

```python
# El primer parámatro es la lista de los valores a representar. Puede ser una columna del DF ya calculada o una variable tipo lista que los contenga. 
plt.pie(titxsex.PassengerId, labels = titxsex.Sex, colors = ['pink', 'cyan'],
        autopct='%1.1f%%', shadow=True, explode = (0.20 , 0), startangle = 0) 
#autopct muestra % dentro
#explode separa los pedazos del chart (en relacion al centro)
#start angle (rotacion del chart)
plt.title = ("TITANIC  - Male / Female proportion'")
plt.tight_layout()
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
# si en el xlim y el y lim se colocan valores negativos despega la data del eje (como en el ejemplo)
sns.set_style(style="darkgrid")

sns.FacetGrid(titanic, margin_titles=True, height=6, legend_out=True, ylim = (-20,600), xlim = (-5,85), palette = "muted")
ax = sns.scatterplot(x="Age", y="Fare", hue="Sex",
             alpha=.5,
            data=titanic)
ax.set_title("Fare x Age")

plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1, borderaxespad=-1.2)

plt.show()
```

```python
# Usando SEABORN
sns.set_style(style="darkgrid")
ax = sns.distplot(titanic['Fare'],  
            color = 'blue', 
            bins = 100, 
            kde =False)
ax.set_title("Fare Price Histogram")
ax.set_ylabel("Frequency")

plt.show()
```
