# Tips

```python
import pandas as pd

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns


# print the graphs in the notebook
%matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
dset = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv")
```

### Step 3. Assign it to a variable called tips

```python
tips = dset
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
tips.hist("total_bill", bins=20)
```

```python
sns.distplot(tips['total_bill'], color = 'purple')
```

```python
sns.distplot(tips['total_bill'], color = 'purple', bins = 20, kde =False) #Mas bins y elim la density line 
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tips.plot.scatter (  x='total_bill',
                    y='tip',
                    title = ('Scatter plot of total_bill and tip'),
                    c = 'b',                     # Color de los puntos segun el sexo
                    alpha = 0.3                     # Para la transparencia de los puntos
                )
```

```python
sns.relplot(x="total_bill", y="tip", 
            sizes=(40, 400), alpha=.5, palette="muted",
            data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", 
            sizes=(40, 400), alpha=.5, palette="deep",
            data=tips)
sns.jointplot("total_bill", "tip", data=tips, kind='scatter')  #Juntando Joint distribution
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.catplot(x="day", y="total_bill", hue="time",
             alpha=1, palette="deep",
            data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
             alpha=1, palette="deep",
            data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex",
             alpha=1, palette="deep",
            data=tips)
```

```python
sns.catplot(x="day", y="total_bill",
             alpha=1, palette="deep",
            data=tips)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.catplot(y="day", x="tip", hue= "sex",
             alpha=0.8, palette="deep",
            data=tips)
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.catplot(x="day", y="total_bill",hue="time", kind="box", data=tips);
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.FacetGrid(tips, col="time").map(plt.hist, "tip")
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
sns.FacetGrid(tips, col="sex", hue="smoker").map(plt.scatter, "total_bill", "tip", alpha = 0.5).add_legend()
```
