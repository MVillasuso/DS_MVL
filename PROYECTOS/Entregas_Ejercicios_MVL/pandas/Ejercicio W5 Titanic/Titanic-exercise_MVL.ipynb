34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True

![kdeplot](kdeplotsur.png)

```python
tit_deads = titanic[titanic.survived == 0]
tit_deads.shape
```

```python
tit_deads.info()
```

```python
tit_deads.drop(columns= ["survived", "name", "sibsp", "parch", "ticket", "fare", "cabin", "embarked", "boat", "body", "home.dest"], inplace=True)
```

```python
tit_deads.head(5)
```

```python
display(tit_deads)
```

```python
tit_deads[tit_deads.age.isna()].count()
```

```python
#Reemplace todas las edades null por cero
# tit_deads.fillna(0,inplace=True)
```

```python
# o borrando las filas donde las edades sean cero
tit_deads.dropna(axis=0, how='any', inplace = True)
```

```python
tit_deads.info()
```

```python
#your code
sns.set(style="whitegrid")
sns.kdeplot(data= tit_deads.age,color = "red", label = "Not Survived",shade = True)
```

```python
#Por edad y clase (Los fallecidos de primera clase eran mayores)
fg = sns.FacetGrid(tit_deads, hue="pclass", aspect=3)
fg.map(sns.kdeplot, "age", shade=True)
```

```python
# breakdown by age and sex
fg = sns.FacetGrid(titanic, col="gender", row="pclass", hue="gender", size=2.5, aspect=2.5)
fg.map(sns.kdeplot, "age", shade=True)
```
