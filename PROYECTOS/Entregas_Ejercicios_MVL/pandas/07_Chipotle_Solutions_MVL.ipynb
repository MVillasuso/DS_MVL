```python
dset = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv","\t")
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = dset
chipo.head(10)
```

### Step 4. See the first 10 entries

### Step 5. Create a histogram of the top 5 items bought

```python
top5 = chipo.groupby("item_name").sum()["quantity"].sort_values(ascending= False).index[0:5].to_list()
```

```python
top5
```

```python
chipo.head(5)
```

```python
df_top5 = chipo[(chipo.item_name.isin (top5))]
```

```python
df_top5.shape
```

```python
df_top5.head(5)
```

```python
#HISTOGRAMA POR CANTIDAD (SIN AGRUPAR POR ITEM_NAME)
df_top5.hist("quantity")
```

```python
# HISTOGRAMA POR CANTIDAD TOTAL VENDIDA (BINS) (AGRUPADA POR ITEMS MAS VENDIDOS)
df_top5.groupby("item_name").sum()["quantity"].to_frame().hist()
```

```python
df_top5.groupby("item_name").sum()["quantity"].to_frame()
```

```python
serbyitem=df_top5.groupby("item_name").sum()["quantity"].sort_values(ascending = False).to_frame()
serbyitem
```

```python
# GRAFICO DE BARRAS. POR CANTIDAD VENDIDA (ITEMS MAS VENDIDOS)
serbyitem.plot(kind='bar')
plt.xlabel ("Item")
plt.ylabel ("Quantity")
plt.title ("Top 5 items bought")
plt.show()
```

```python
serbyitem
```

```python
# Otra forma usando MatplotLib directamente
plt.bar(serbyitem.index, serbyitem.quantity, label = 0)
plt.xlabel ("Items")
plt.ylabel ("Quantity")
plt.title ("Top 5 Items Chipotle")
plt.xticks(rotation = "vertical")
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price2'] = chipo['item_price'].str[1:].astype(float)
chipo['tot_price'] = chipo['item_price2'] * chipo["quantity"]
```

```python
chipo.head(10)
```

```python
dfx = chipo.groupby("order_id").sum()
```

```python
dfx.sort_values(by=["tot_price"], ascending = False)
```

```python
chipo[chipo["order_id"]==1443]
```

```python
dfx.plot.scatter(x="item_price2", y="quantity", title= "Scatter plot between item price and quantity");
plt.show(block=True)
```

```python
dfx.plot.scatter(x="tot_price", y="quantity", title= "Scatter plot between total price and quantity");
plt.show(block=True)
```
