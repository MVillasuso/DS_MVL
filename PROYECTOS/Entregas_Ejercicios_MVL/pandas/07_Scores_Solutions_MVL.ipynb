# Scores

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],  
        'age': [42,52,36,24,73],
        'female': [0,1,1,0,1],
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}
```

```python
reg = pd.DataFrame(raw_data)
```

```python
reg
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
reg.plot.scatter (x='preTestScore',
                    y='postTestScore',
                    title = ('Scatter plot of preTestScore and postTestScore by age'),
                    c ='red',           # Color de los puntos
                    s = reg.age)        #Vara el tamano del punto en base a la edad
plt.show(block=True)
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sex_val =lambda x: 'F' if x else 'M'
reg['sex'] = reg['female'].apply(sex_val)
```

```python
reg
```

```python
colorv = reg['sex'].map({'M': 'b', 'F': 'r'})
reg.plot.scatter (  x='preTestScore',
                    y='postTestScore',
                    title = ('Scatter plot of preTestScore and postTestScore by age'),
                    c = colorv,                        # Color de los puntos segun el sexo
                    alpha = 0.3,                        # Para la transparencia de los puntos
                    s = reg.postTestScore * 4.5)        #Vara el tamano del punto en base a postTestScore
plt.show(block=True)
```
